```python
# Hypothesis testing 
```

```python
# 1st Test: Anova Test
'''
Does revenue differ across customer segments?

Ho: Revenue is equal & does not differ across customer segments
H1: Revenue is not equal & differs across customer segments
'''
```

```python
df['customer_segment'].value_counts()
```

```python
returning=df[df['customer_segment']=='Returning']['total_revenue']
loyal=df[df['customer_segment']=='Loyal']['total_revenue']
new=df[df['customer_segment']=='New']['total_revenue']
```

```python
from scipy.stats import f_oneway
f_stat,p_value=f_oneway(returning,loyal,new)
print('Fstat:',f_stat)
print('P-value:',p_value)
```

```python
df.groupby('customer_segment')['total_revenue'].mean()
```

```python
'''
Here, p-value:0.71 which is >0.05 so we fail to reject the null hypothesis, so it can be stated that there is no statistical evidence that says that revenue differs across 
customer segments which get validated by getting the average reveune after grouping by customer segments: Loyal=8257.79, New=8007.62, Returning=8265.98 &
statistically the difference is very small to reject the null hypothesis and say that revenue is not equal and differs across segments
'''
```

```python
# 2nd Test: Independent T-Test

'''
Does discount affect revenue?

Ho: There is no difference in revenue after discounts applied
H1: There is a difference in revenue after discounts applied
'''
```

```python
df.head()
```

```python
df['discount_offered'].value_counts()
```

```python
df['discount_applied']=df['discount_offered'].apply(lambda x:0 if x=='No Discount' else 1)
```

```python
with_discount=df[df['discount_applied']==1]['total_revenue']
without_discount=df[df['discount_applied']==0]['total_revenue']
```

```python
from scipy.stats import ttest_ind
t_stat,p_value=ttest_ind(with_discount,without_discount)
print('P-value:',p_value)
```

```python
df.groupby('discount_applied')['total_revenue'].mean()
```

```python
'''
Here, p-value(2.696828565523088e-07) which is <0.05 we reject the null hypothesis, hence statistical evidence states that there is a difference in 
revenue after discounts applied which can be validated while getting the average revenue after grouping it by discount applied group(7764) and no
discount applied group(9465) ie the orders where no discount is being applied is contributing more to the total revenue
'''
```

```python
# purchase prediction model
```

```python
df.head()
```

```python
df.columns
```

```python
x=df[['age','gender','region','customer_segment','category','discount_offered','marketing_channel','review_score',
      'discount_pct','month','quarter']]
y=df['purchase_made']
```

```python
x_encoded = pd.get_dummies(x, columns=['gender', 'region', 'customer_segment','discount_offered',
                                        'category',
                                        'marketing_channel'],
                            drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_encoded,y,test_size=0.2,random_state=42)
```

```python
#Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
```

```python
lr_model.fit(x_train,y_train)
```

```python
lr_ypred=lr_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print('Accuracy_score:',accuracy_score(y_test,lr_ypred))
print('Classification report:',classification_report(y_test,lr_ypred))
print('Confusion Matrix:',confusion_matrix(y_test,lr_ypred))
```

```python
df['purchase_made'].value_counts()
```

```python
# Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
rf_model.fit(x_train,y_train)
```

```python
rf_ypred=rf_model.predict(x_test)
```

```python
print('Accuracy_score:',accuracy_score(y_test,rf_ypred))
print('Classification report:',classification_report(y_test,rf_ypred))
print('Confusion Matrix:',confusion_matrix(y_test,rf_ypred))
```

```python
# Tuned Random Forest Model
```

```python
rf_bal=RandomForestClassifier(n_estimators=200,max_depth=8,min_samples_split=10,class_weight='balanced', random_state=42)
rf_bal.fit(x_train,y_train)
```

```python
rfbal_ypred=rf_bal.predict(x_test)
```

```python
print('Accuracy_score:',accuracy_score(y_test,rfbal_ypred))
print('Classification report:',classification_report(y_test,rfbal_ypred))
print('Confusion Matrix:',confusion_matrix(y_test,rfbal_ypred))
```

```python
# With Probability Threshold Tuning
```

```python
y_prob=rf_bal.predict_proba(x_test)[:,1]
y_pred_prob = (y_prob > 0.3).astype(int)
```

```python
print('Accuracy_score:',accuracy_score(y_test,y_pred_prob))
print('Classification report:',classification_report(y_test,y_pred_prob))
print('Confusion Matrix:',confusion_matrix(y_test,y_pred_prob))
```

```python
'''

Instead of maximizing accuracy, I optimized the model for business impact. By lowering the threshold, the model captures 94% of potential buyers, which
is more valuable in a marketing context even at the cost of precision

So, Random Forest with probability based threshold tuning was chosen as the final model due to its ability to capture potential buyers.
'''
```
